```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('derm.csv')
dataset
```

```python
X = dataset.iloc[:, 0:34].values
y = dataset.iloc[:, 34].values
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X = sc_X.fit_transform(X)
X
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 5)
X = pca.fit_transform(X)
explained_variance = pca.explained_variance_ratio_
explained_variance
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

```python
from sklearn.linear_model import LogisticRegression
logisticregressionclassifier = LogisticRegression(random_state = 0)
logisticregressionclassifier.fit(X_train, y_train)
```

```python
logisticregression_y_pred = logisticregressionclassifier.predict(X_test)
logisticregression_y_pred
```

```python
from sklearn.metrics import confusion_matrix
logisticregression_cm = confusion_matrix(y_test, logisticregression_y_pred)
logisticregression_cm
```

```python
from sklearn.tree import DecisionTreeClassifier
decisiontreeclassifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
decisiontreeclassifier.fit(X_train, y_train)
```

```python
decisiontree_y_pred = decisiontreeclassifier.predict(X_test)
decisiontree_y_pred
```

```python
decisiontree_cm = confusion_matrix(y_test, decisiontree_y_pred)
decisiontree_cm
```

```python
from sklearn.ensemble import RandomForestClassifier
randomforestclassifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
randomforestclassifier.fit(X_train, y_train)
```

```python
randomforest_y_pred = randomforestclassifier.predict(X_test)
randomforest_y_pred
```

```python
randomforest_cm = confusion_matrix(y_test, randomforest_y_pred)
randomforest_cm
```

```python
from sklearn.svm import SVC
svmclassifier = SVC(kernel = 'linear', random_state = 0)
svmclassifier.fit(X_train, y_train)
```

```python
svm_y_pred = svmclassifier.predict(X_test)
svm_y_pred
```

```python
svm_cm = confusion_matrix(y_test, svm_y_pred)
svm_cm
```
